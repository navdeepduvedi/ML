```python
py.hist(x='instant' ,data=data,bins=80)
py.grid()
py.title('Count plot of instant in day')
```

```python
py.figure(figsize=(10,6))
sns.regplot(x='weekday',y='holiday',data=data , x_jitter=0.3 , scatter_kws={'alpha':0.2})
sns.despine()
```

```python
py.figure(figsize=(10,6))
sns.regplot(x='weekday',y='holiday',data=hours , x_jitter=0.3 , scatter_kws={'alpha':0.2})
sns.despine()
```
